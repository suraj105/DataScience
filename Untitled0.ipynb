```python
df=pd.read_csv('abc.csv')
df #before changing the base index
```

```python
df = df.set_index(pd.DatetimeIndex(df['Date'].values)) #setting index as date as we make change in order to the date
df #after chaging the base index
```

```python
#calculating SMA, SD, UB and LB

#time period of 20, depends on how compact you want the interval to be 
#Simple moving average, Standard Deviation, Upper Band and Lower Band

period = 20

# According to the Bollinger Band , The following formulas are to be implemented
#Middle Band = n-period moving average
#Upper Band = Middle Band + ( y * n-period standard deviation)
#Lower Band = Middle Band - ( y * n-period standard deviation)
#Where:
#    n = number of periods
#    y = factor to apply to the standard deviation value, (typical default for y = 2)




#for SMA
df['SMA'] = df['Close'].rolling(window=period).mean()
#for SD
df['SD'] = df['Close'].rolling(window=period).std()
# Upper Bollinger Band
df['Upper'] = df['SMA'] + (df['SD'] * 2)
# lower Bollinger band
df['Lower'] = df['SMA'] - (df['SD'] * 2)

#create a list of Column with added Close , SD, Upper and Lower Data
column_list =['Close', 'SD', 'Upper','Lower']
#ploting the given data in chart
df[column_list].plot(figsize=(20,10)) #determining figure size

import matplotlib.pyplot as pot

pot.title('Bollinger Band Strategy for Bitcoin') #title for the Chart
pot.ylabel('USD Price in $') # labeling the y axis of the given Chart
pot.xlabel('Date in years') # labeling the x axis of the given Chart
pot.yticks(rotation = 15) #rotation of price by 15 degree 
pot.show() #Display the  Graph
```

```python
#plot shade area between Bollinger bands

fig = pot.figure(figsize=(20,8))
#adding sub plot for gray area
gray = fig.add_subplot(111) # 
#get iindex values of data frame
x_axis = df.index
#plot and shade the area between UP and LB
gray.fill_between(x_axis, df['Upper'],df['Lower'],color ='grey')
#plot closing price and average
gray.plot(x_axis, df['Close'], color ='green', lw =2, label='Close Price')
gray.plot(x_axis, df['SMA'], color ='blue', lw =2, label='Simple moving average')
gray.set_title('Bollinger Band for Bitcoin')
gray.set_xlabel('Date in years')
gray.set_ylabel('USD Price in $')
pot.yticks(rotation = 15) #rotation of price by 15 degree 
gray.legend() #to sshow the legend
pot.show()  # to show the chart
```

```python
#create new data

new_df = df[period-1:]
new_df
```

```python
def get_signal(data):

  buy_signal = []
  sell_signal = []

  for i in range(len(data['Close'])):
    if data['Close'][i] > data['Upper'][i]: # then you should sell
      buy_signal.append(np.nan)
      sell_signal.append(data['Close'][i])
    elif data['Close'][i] < data['Lower'][i]: # then buy
      buy_signal.append(data['Close'][i])
      sell_signal.append(np.nan)
    else:
        buy_signal.append(np.nan)
        sell_signal.append(np.nan)

  return(buy_signal, sell_signal)
```

```python
import numpy as np

#Create two new columns
new_df['Buy'] = get_signal(new_df)[0]
new_df['Sell'] = get_signal(new_df)[1]

```

```python
#plot all the data

fig = pot.figure(figsize=(30,15)) #size of graph
#add plot
gray = fig.add_subplot(1,1,1)
#get iindex values of data frame
x_axis = new_df.index
#plot and shade the area between UP and LB
gray.fill_between(x_axis, new_df['Upper'],new_df['Lower'],color ='grey')
#plot closing price and average
gray.plot(x_axis, new_df['Close'], color ='red', lw =3, label='Close Price', alpha= 0.1)
gray.plot(x_axis, new_df['SMA'], color ='blue', lw =3, label='Simple moving average', alpha= 0.1)
gray.scatter(x_axis, new_df['Buy'], color ='green', lw=3 , label='Buy', marker='^', alpha= 1) #adding the green buy arrow , when the algorith tell us to buy
gray.scatter(x_axis, new_df['Sell'], color ='red', lw=3 , label='Sell', marker='v', alpha= 1) #adding the red sell arrow , when the algorith tell us to sell

gray.set_title('Bollinger Band for Bitcoin')
gray.set_xlabel('Date')
gray.set_ylabel('USD Price in $')
pot.yticks(rotation = 15) #rotation of price by 15 degree 
gray.legend()  #showing the legend
pot.show() #displaying the graph
```
